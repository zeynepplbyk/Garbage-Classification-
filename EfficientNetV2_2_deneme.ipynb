# ADAM +CROSSENTROPY

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from tqdm import tqdm
import copy
import os
import shutil
from google.colab import drive

# ---------------------------------------------------------
# 1. DRIVE BAĞLANTISI VE HIZLANDIRMA
# ---------------------------------------------------------
if not os.path.exists('/content/drive'):
    print("🔌 Google Drive bağlanıyor...")
    drive.mount('/content/drive')

# Orijinal Drive Yolu
DRIVE_PATH = "/content/drive/MyDrive/Derin_Ogrenme/merged_dataset2"

# Colab'in kendi hızlı diski (Local Runtime)
LOCAL_PATH = "/content/temp_dataset"

print(f"🚀 HIZLANDIRMA BAŞLIYOR: Veriler Drive'dan yerel diske kopyalanıyor...")
if not os.path.exists(LOCAL_PATH):
    shutil.copytree(DRIVE_PATH, LOCAL_PATH)
    print("✅ Kopyalama tamamlandı! Eğitim başlıyor.")
else:
    print("✅ Veriler zaten kopyalanmış, devam ediliyor.")

# Klasör Yolları (Yerel Diskten)
TRAIN_DIR = os.path.join(LOCAL_PATH, 'train')
VAL_DIR = os.path.join(LOCAL_PATH, 'val')
TEST_DIR = os.path.join(LOCAL_PATH, 'test')

# ---------------------------------------------------------
# 2. AYARLAR
# ---------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"⚙️ Çalışılan Cihaz: {device}")

BATCH_SIZE = 32
EPOCHS = 20
LEARNING_RATE = 1e-4

# ---------------------------------------------------------
# 3. VERİ YÜKLEME (SADE - AUGMENTATION YOK)
# ---------------------------------------------------------
norm_mean = [0.485, 0.456, 0.406]
norm_std = [0.229, 0.224, 0.225]

base_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(norm_mean, norm_std)
])

print("📂 Veriler yükleniyor...")
train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=base_transforms)
val_dataset = datasets.ImageFolder(VAL_DIR, transform=base_transforms)
test_dataset = datasets.ImageFolder(TEST_DIR, transform=base_transforms)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

class_names = train_dataset.classes
print(f"✅ Sınıflar: {class_names}")

# ---------------------------------------------------------
# 4. MODEL (EFFICIENTNET V2)
# ---------------------------------------------------------
model = models.efficientnet_v2_s(weights=models.EfficientNet_V2_S_Weights.DEFAULT)
in_features = model.classifier[1].in_features
model.classifier[1] = nn.Linear(in_features, len(class_names))
model = model.to(device)

# ---------------------------------------------------------
# 5. OPTIMIZER & LOSS (İstediğin Ayarlar)
# ---------------------------------------------------------
# 1. OPTIMIZER: ADAM
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

# 2. LOSS: STANDART CROSS ENTROPY (Label Smoothing YOK)
criterion = nn.CrossEntropyLoss()

# Scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3)

# ---------------------------------------------------------
# 6. EĞİTİM DÖNGÜSÜ
# ---------------------------------------------------------
best_model_wts = copy.deepcopy(model.state_dict())
best_acc = 0.0
patience_counter = 0
PATIENCE_LIMIT = 5

history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

print(f"\n🔥 EĞİTİM BAŞLIYOR (Adam + Standard CrossEntropy)...")

for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    # --- TRAIN ---
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    loop = tqdm(train_loader, leave=True)
    for inputs, labels in loop:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += torch.sum(preds == labels.data)
        total += labels.size(0)
        loop.set_description(f"Train Loss: {loss.item():.4f}")

    epoch_loss = running_loss / len(train_dataset)
    epoch_acc = correct.double() / len(train_dataset)
    history['train_loss'].append(epoch_loss)
    history['train_acc'].append(epoch_acc.item())

    # --- VAL ---
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += torch.sum(preds == labels.data)

    val_epoch_loss = val_loss / len(val_dataset)
    val_epoch_acc = val_correct.double() / len(val_dataset)
    history['val_loss'].append(val_epoch_loss)
    history['val_acc'].append(val_epoch_acc.item())

    print(f"Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}")
    print(f"Val Loss:   {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f}")

    scheduler.step(val_epoch_loss)

    # --- SAVE BEST & EARLY STOPPING ---
    # Modeli Drive'a kaydet (Kalıcı olsun)
    save_path = '/content/drive/MyDrive/Derin_Ogrenme/best_garbage_model_adam_ce.pth'

    if val_epoch_acc > best_acc:
        best_acc = val_epoch_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        torch.save(model.state_dict(), save_path)
        print(f"✅ Yeni en iyi skor! Model kaydedildi: {save_path}")
        patience_counter = 0
    else:
        patience_counter += 1
        print(f"⚠️ İyileşme yok ({patience_counter}/{PATIENCE_LIMIT})")

    if patience_counter >= PATIENCE_LIMIT:
        print("⏹️ Early Stopping.")
        break

# ---------------------------------------------------------
# 7. DEĞERLENDİRME
# ---------------------------------------------------------
model.load_state_dict(best_model_wts)

# Grafikler
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history['train_acc'], label='Train')
plt.plot(history['val_acc'], label='Val')
plt.title('Accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history['train_loss'], label='Train')
plt.plot(history['val_loss'], label='Val')
plt.title('Loss (Standard CE)')
plt.legend()
plt.show()

# Test Raporu
print("\n📊 TEST SETİ RAPORU...")
all_preds = []
all_labels = []
model.eval()
with torch.no_grad():
    for inputs, labels in tqdm(test_loader):
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

print(classification_report(all_labels, all_preds, target_names=class_names))

cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.ylabel('Gerçek')
plt.xlabel('Tahmin')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from tqdm.notebook import tqdm  # Notebook için daha uyumlu tqdm
import copy
import os
import shutil
import time
from concurrent.futures import ThreadPoolExecutor
from google.colab import drive

# ---------------------------------------------------------
# 1. DRIVE BAĞLANTISI VE HIZLI KOPYALAMA AYARLARI
# ---------------------------------------------------------
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# Yollar
SOURCE_DIR = "/content/drive/MyDrive/Derin_Ogrenme/merged_dataset2"  # Drive'daki kaynak
DEST_DIR = "/content/temp_dataset_fast"                             # Colab'deki hedef (Hızlı Disk)

print(f"Kaynak: {SOURCE_DIR}")
print(f"Hedef : {DEST_DIR}")

# ---------------------------------------------------------
# 2. PARALEL KOPYALAMA FONKSİYONLARI (HIZLANDIRICI)
# ---------------------------------------------------------
def copy_file(src, dst):
    # Hedef klasör yoksa oluştur
    os.makedirs(os.path.dirname(dst), exist_ok=True)
    # Dosyayı kopyala
    shutil.copy2(src, dst)

def copy_folder_parallel(src_folder, dest_folder, num_threads=16):
    files_to_copy = []

    # Tüm dosya yollarını topla
    print("Dosya listesi oluşturuluyor...")
    for root, dirs, files in os.walk(src_folder):
        for file in files:
            src_file = os.path.join(root, file)
            # Hedef yolu oluştur
            rel_path = os.path.relpath(src_file, src_folder)
            dest_file = os.path.join(dest_folder, rel_path)
            files_to_copy.append((src_file, dest_file))

    print(f"Toplam {len(files_to_copy)} dosya kopyalanacak. İşlem başlıyor...")

    # Paralel kopyalama işlemi (ThreadPoolExecutor ile)
    start_time = time.time()
    with ThreadPoolExecutor(max_workers=num_threads) as executor:
        # Tqdm ile ilerleme çubuğu göster
        list(tqdm(executor.map(lambda p: copy_file(*p), files_to_copy), total=len(files_to_copy)))

    duration = time.time() - start_time
    print(f"Kopyalama tamamlandı! Süre: {duration:.2f} saniye")

# --- KOPYALAMA İŞLEMİNİ BAŞLAT ---
if os.path.exists(DEST_DIR):
    print("Veri seti zaten var, kopyalama atlanıyor (Temiz kurulum istersen bu klasörü sil).")
else:
    copy_folder_parallel(SOURCE_DIR, DEST_DIR)

# Veri yollarını güncelle
TRAIN_DIR = os.path.join(DEST_DIR, 'train')
VAL_DIR = os.path.join(DEST_DIR, 'val')
TEST_DIR = os.path.join(DEST_DIR, 'test')

# ---------------------------------------------------------
# 3. AYARLAR (İLK KODDAN AYNEN ALINDI)
# ---------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
BATCH_SIZE = 32
EPOCHS = 20
LEARNING_RATE = 1e-4

# ---------------------------------------------------------
# 4. VERİ YÜKLEME (Augmentation Yok - Sade)
# ---------------------------------------------------------
norm_mean = [0.485, 0.456, 0.406]
norm_std = [0.229, 0.224, 0.225]

base_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(norm_mean, norm_std)
])

train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=base_transforms)
val_dataset = datasets.ImageFolder(VAL_DIR, transform=base_transforms)
test_dataset = datasets.ImageFolder(TEST_DIR, transform=base_transforms)

# num_workers=2 veri yerelde olduğu için performanslı çalışır
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

class_names = train_dataset.classes
print(f"Sınıflar: {class_names}")

# ---------------------------------------------------------
# 5. MODEL (EfficientNetV2 + Dropout - İLK KODDAN)
# ---------------------------------------------------------
model = models.efficientnet_v2_s(weights=models.EfficientNet_V2_S_Weights.DEFAULT)

# EfficientNet'in classifier katmanını özelleştiriyoruz
in_features = model.classifier[1].in_features

# DAHA GÜÇLÜ BİR CLASSIFIER (Senin isteğin üzerine)
model.classifier = nn.Sequential(
    nn.Dropout(p=0.5),                    # %50 ihtimalle nöronları kapat (Ezberi bozar)
    nn.Linear(in_features, len(class_names))
)

model = model.to(device)

# ---------------------------------------------------------
# 6. OPTIMIZER (Weight Decay - İLK KODDAN)
# ---------------------------------------------------------
# weight_decay=1e-4: Modele "çok karmaşıklaşma" cezası verir.
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-4)

# Loss Fonksiyonu (Standard CrossEntropy - İLK KODDAN)
criterion = nn.CrossEntropyLoss()

scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3)

# ---------------------------------------------------------
# 7. EĞİTİM DÖNGÜSÜ
# ---------------------------------------------------------
best_model_wts = copy.deepcopy(model.state_dict())
best_acc = 0.0
patience_counter = 0
PATIENCE_LIMIT = 5

history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

print(f"\n EĞİTİM BAŞLIYOR (EfficientNetV2 | Dropout + Weight Decay Aktif)...")

for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    # --- TRAIN ---
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    loop = tqdm(train_loader, leave=True)
    for inputs, labels in loop:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += torch.sum(preds == labels.data)
        total += labels.size(0)
        loop.set_description(f"Train Loss: {loss.item():.4f}")

    epoch_loss = running_loss / len(train_dataset)
    epoch_acc = correct.double() / len(train_dataset)
    history['train_loss'].append(epoch_loss)
    history['train_acc'].append(epoch_acc.item())

    # --- VAL ---
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += torch.sum(preds == labels.data)

    val_epoch_loss = val_loss / len(val_dataset)
    val_epoch_acc = val_correct.double() / len(val_dataset)
    history['val_loss'].append(val_epoch_loss)
    history['val_acc'].append(val_epoch_acc.item())

    print(f"Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}")
    print(f"Val Loss:   {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f}")

    scheduler.step(val_epoch_loss)

    # --- SAVE ---
    # Model ismi orijinal koddaki gibi bırakıldı
    save_path = '/content/drive/MyDrive/Derin_Ogrenme/best_garbage_model_adam_ce_dropout.pth'

    if val_epoch_acc > best_acc:
        best_acc = val_epoch_acc
        best_model_wts = copy.deepcopy(model.state_dict())
        torch.save(model.state_dict(), save_path)
        print(f" Yeni en iyi skor! Kaydedildi.")
        patience_counter = 0
    else:
        patience_counter += 1
        print(f" İyileşme yok ({patience_counter}/{PATIENCE_LIMIT})")

    if patience_counter >= PATIENCE_LIMIT:
        print(" Early Stopping.")
        break

# ---------------------------------------------------------
# 8. SONUÇLAR VE TEST
# ---------------------------------------------------------
model.load_state_dict(best_model_wts)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history['train_acc'], label='Train')
plt.plot(history['val_acc'], label='Val')
plt.title('Accuracy (EfficientNetV2)')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history['train_loss'], label='Train')
plt.plot(history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()
plt.show()

# Test Raporu
all_preds = []
all_labels = []
model.eval()
print("\nTest seti üzerinde tahminler alınıyor...")
with torch.no_grad():
    for inputs, labels in tqdm(test_loader):
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

print("\n--- Classification Report ---")
print(classification_report(all_labels, all_preds, target_names=class_names))

cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.ylabel('Gerçek')
plt.xlabel('Tahmin')
plt.show()
```

# overfittingi önlemek icin;Backbone Freezing (Gövdeyi Dondurma)+Batch Normalization Eklendi+Dropout Artırıldı



```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from tqdm import tqdm
import copy
import os
import shutil
from google.colab import drive
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
# ---------------------------------------------------------
# 1. DRIVE ve HIZLANDIRMA
# ---------------------------------------------------------
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

DRIVE_PATH = "/content/drive/MyDrive/Derin_Ogrenme/merged_dataset2"
LOCAL_PATH = "/content/temp_dataset"

if not os.path.exists(LOCAL_PATH):
    print(" Veriler kopyalanıyor...")
    shutil.copytree(DRIVE_PATH, LOCAL_PATH)
else:
    print(" Veriler hazır.")

TRAIN_DIR = os.path.join(LOCAL_PATH, 'train')
VAL_DIR = os.path.join(LOCAL_PATH, 'val')
TEST_DIR = os.path.join(LOCAL_PATH, 'test')
# --- BU KODU 2. BÖLÜMDEN ÖNCE ÇALIŞTIR ---
import os
import shutil

def temizle_ve_kontrol_et(path):
    # .ipynb_checkpoints gibi gizli klasörleri siler
    for root, dirs, files in os.walk(path):
        for d in dirs:
            if d.startswith('.'):
                shutil.rmtree(os.path.join(root, d))
                print(f"Silindi: {os.path.join(root, d)}")

print("Veri seti temizleniyor...")
temizle_ve_kontrol_et(LOCAL_PATH)

# Sınıfları tekrar kontrol et
temp_train_data = datasets.ImageFolder(TRAIN_DIR)
temp_val_data = datasets.ImageFolder(VAL_DIR)

print(f"\nTrain Sınıfları ({len(temp_train_data.classes)}): {temp_train_data.classes}")
print(f"Val Sınıfları   ({len(temp_val_data.classes)}): {temp_val_data.classes}")

if temp_train_data.classes != temp_val_data.classes:
    raise ValueError("HATA: Train ve Val klasörlerindeki sınıf isimleri veya sayıları eşit değil! Lütfen klasörleri kontrol et.")
else:
    print("\n✅ BAŞARILI: Sınıflar eşit. Eğitime geçebilirsin.")
# ---------------------------------------------------------
# 2. AYARLAR
# ---------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
BATCH_SIZE = 32
EPOCHS = 25  # Early stopping olduğu için epoch'u biraz artırabiliriz
LEARNING_RATE = 1e-4

# ---------------------------------------------------------
# 3. VERİ YÜKLEME (Sade - Augmentation Yok)
# ---------------------------------------------------------
norm_mean = [0.485, 0.456, 0.406]
norm_std = [0.229, 0.224, 0.225]

base_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(norm_mean, norm_std)
])

train_dataset = datasets.ImageFolder(TRAIN_DIR, transform=base_transforms)
val_dataset = datasets.ImageFolder(VAL_DIR, transform=base_transforms)
test_dataset = datasets.ImageFolder(TEST_DIR, transform=base_transforms)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

class_names = train_dataset.classes

# ---------------------------------------------------------
# 4. MODEL (Overfitting Önleyici Revizyonlar)
# ---------------------------------------------------------
model = models.efficientnet_v2_s(weights=models.EfficientNet_V2_S_Weights.DEFAULT)

# [KRİTİK ADIM] Feature Extractor (Gövde) Kısmını Donduruyoruz
# Bu, modelin devasa parametrelerinin ezber yapmasını engeller.
for param in model.features.parameters():
    param.requires_grad = False

# Classifier Katmanını Özelleştirme
in_features = model.classifier[1].in_features

# İSTEDİĞİN ÜÇLÜ BURADA: Batch Norm + Dropout + Yapısal Düzenleme
model.classifier = nn.Sequential(
    nn.BatchNorm1d(in_features),      # [Batch Norm] Gelen özellikleri normalize et
    nn.Linear(in_features, 512),      # Ara katman
    nn.ReLU(),
    nn.Dropout(p=0.6),                # [Dropout] %60 kapatma (Agresif Regularization)
    nn.Linear(512, len(class_names))  # Çıkış katmanı
)

model = model.to(device)

# ---------------------------------------------------------
# 5. OPTIMIZER
# ---------------------------------------------------------
# Sadece classifier parametrelerini optimize edeceğiz (Gövde donuk olduğu için)
optimizer = optim.Adam(model.classifier.parameters(), lr=LEARNING_RATE, weight_decay=1e-3)
# Not: weight_decay'i 1e-4'ten 1e-3'e çektim, cezayı artırdık.

criterion = nn.CrossEntropyLoss()
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=3)

# ---------------------------------------------------------
# 6. EĞİTİM (Early Stopping Dahil)
# ---------------------------------------------------------
best_model_wts = copy.deepcopy(model.state_dict())
best_loss = float('inf') # Loss üzerinden takip etmek overfitting için daha güvenlidir
patience_counter = 0
PATIENCE_LIMIT = 5 # [Early Stopping] Sabır sınırı

history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}

print(f"\n EĞİTİM BAŞLIYOR (Layer Freezing + BN + Dropout Aktif)...")

for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch+1}/{EPOCHS}")

    # --- TRAIN ---
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    loop = tqdm(train_loader, leave=True)
    for inputs, labels in loop:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += torch.sum(preds == labels.data)
        total += labels.size(0)
        loop.set_description(f"Train Loss: {loss.item():.4f}")

    epoch_loss = running_loss / len(train_dataset)
    epoch_acc = correct.double() / len(train_dataset)
    history['train_loss'].append(epoch_loss)
    history['train_acc'].append(epoch_acc.item())

    # --- VAL ---
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(outputs, 1)
            val_correct += torch.sum(preds == labels.data)

    val_epoch_loss = val_loss / len(val_dataset)
    val_epoch_acc = val_correct.double() / len(val_dataset)
    history['val_loss'].append(val_epoch_loss)
    history['val_acc'].append(val_epoch_acc.item())

    print(f"Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}")
    print(f"Val Loss:   {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f}")

    scheduler.step(val_epoch_loss)

    # --- EARLY STOPPING CHECK (Val Loss'a göre) ---
    # Overfitting kontrolü için Val Loss'un düşmesi Acc'den daha kritiktir.
    if val_epoch_loss < best_loss:
        best_loss = val_epoch_loss
        best_model_wts = copy.deepcopy(model.state_dict())
        torch.save(model.state_dict(), '/content/drive/MyDrive/Derin_Ogrenme/best_garbage_model_regularized.pth')
        print(f" Kaydedildi (Val Loss düştü).")
        patience_counter = 0
    else:
        patience_counter += 1
        print(f" İyileşme yok. Sabır: {patience_counter}/{PATIENCE_LIMIT}")

    if patience_counter >= PATIENCE_LIMIT:
        print(" Early Stopping devreye girdi. Eğitim sonlandırılıyor.")
        break

# ---------------------------------------------------------
# 7. SONUÇLAR
# ---------------------------------------------------------
model.load_state_dict(best_model_wts)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history['train_acc'], label='Train')
plt.plot(history['val_acc'], label='Val')
plt.title('Accuracy (Regularized)')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history['train_loss'], label='Train')
plt.plot(history['val_loss'], label='Val')
plt.title('Loss (Regularized)')
plt.legend()
plt.show()

# Test Raporu
all_preds = []
all_labels = []
model.eval()
with torch.no_grad():
    for inputs, labels in tqdm(test_loader):
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

print(classification_report(all_labels, all_preds, target_names=class_names))
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.ylabel('Gerçek')
plt.xlabel('Tahmin')
plt.show()
```
